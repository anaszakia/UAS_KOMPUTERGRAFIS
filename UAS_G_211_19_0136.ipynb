```python
#upload Dataset curah hujan
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
data1 = pd.read_csv("curah_hujan.csv")
data1.head(37)
```

```python
#membuat multi-bar chart
import matplotlib.pyplot as plt
data1.groupby(['bulan','tahun'])['curah hujan'].sum().unstack().plot(kind='bar')
plt.title('Curah Hujan dari tahun 1981-2010', loc='center', pad=30, fontsize=15, color='blue')
plt.xlabel('Bulan', fontsize= 12)
plt.ylabel('Curah Hujan (mm)', fontsize= 12)
plt.legend(bbox_to_anchor=(1, 1), shadow=True, title='Tahun')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1).astype(int))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#upload Dataset suhu
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
data1 = pd.read_csv("suhu.csv")
data1.head(28)
```

```python
#plot multi-line chart
import matplotlib.pyplot as plt
data1.groupby(['waktu','asimilasi/observasi/non asimilasi'])['suhu'].sum().unstack().plot(marker='.', cmap='plasma')
plt.title('Suhu Permukaan', loc='center', pad=30, fontsize=20, color='blue')
plt.xlabel('Waktu', fontsize=15)
plt.ylabel('Celcius', fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1))
plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)
plt.gcf().set_size_inches(15, 5)
plt.tight_layout()
plt.show()
```
